```python
data_dir = 'data'
os.listdir(data_dir)
```

```python
train_df = pd.read_csv(os.path.join(data_dir, 'train.csv'))
train_df.head()
```

```python
arr_del = []
count = 0
for i in range(len(train_df.index)):
    if count == 1000:
        break
    if train_df.label[i] == 2:
        arr_del.append(i)
        count += 1
```

```python
new_train_df = train_df.drop(train_df.index[arr_del])
```
